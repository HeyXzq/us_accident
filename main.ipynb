```python
import pandas as pd

# Load the dataset
data_df = pd.read_csv("US_Accidents_Dec20_Updated.csv")
data_df.info() # show entries, dtypes, memory useage.

# Have a look
data_df.head(5)
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.describe.html
# Basic statistic on Ordinal, Interval and Ratio data.
OIR_columns = ['Severity', 'Start_Time', 'End_Time', 'Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Distance(mi)', 'Weather_Timestamp', 'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']
data_df[OIR_columns].describe() #Exclude Time; Time can be transformed to second/minute/hour/day/month/year separately, but it is no need to do it here.
```

```python
# Basic statistic on Nominal data
data_df.loc[:, ~data_df.columns.isin(OIR_columns)].astype("object").describe() # All the Nominal data can be treated as "object" type for simplicity.
```
